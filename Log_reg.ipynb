# TEXT PREPROCESSING

```python
# Load spacy english model
nlp = spacy.load("en_core_web_sm", disable = ["parser", "ner"])
```

```python
def clean_text(text):
    # Convert text to a string
    text = str(text)
    # Convert text to lower case to ensure consistency
    text = text.lower()
    # Remove URLs
    text = re.sub(r"https\S+|www\S+|https\S+", "", text)
    # Remove hashtags or mentions
    text = re.sub(r"[@#]\w+", "", text)
    # Remove all html tags
    text = re.sub(r"<.*?>", " ", text)
    # Split numbers in text
    text = re.sub(r"(\d)([A-Za-z])", r"\1 \2", text)   # number before letter
    text = re.sub(r"([A-Za-z])(\d)", r"\1 \2", text)   # number after letter
    # Remove punctuations except basic ones
    text = re.sub(r"[^\w\s']", "", text)
    # Remove extra spaces and tabs
    text = re.sub(r"\s+", " ", text).strip()
    # Expand contractions
    text = con.fix(text)

    doc = nlp(text)
    # Lemmatize, remove stopwords & short tokens
    tokens = [token.lemma_ for token in doc if not token.is_stop and len(token.text) > 2]
    return " ".join(tokens)
```

```python
df['cleaned_reviews'] = df['review'].apply(lambda x: clean_text(x))
```

```python
df.drop(['review' ], axis = 1, inplace = True) 
```

```python
df.head()
```

```python
vectorizer = TfidfVectorizer(stop_words = 'english', max_features = 5000)
```

```python
X = vectorizer.fit_transform(df['cleaned_reviews'])
```

```python
svd = TruncatedSVD(n_components = 1000, random_state = 42)
X_reduced = svd.fit_transform(X)
```

```python
le = LabelEncoder()
df['sentiment'] = le.fit_transform(df[['sentiment']])
```

```python
df['sentiment'].value_counts()
```

```python
y = df['sentiment']
```

# DATA SPLITTING

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size = 0.3, 
                                                   stratify = y, random_state = 42) 
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
%%time
model = LogisticRegression(solver = "saga", max_iter = 1000, n_jobs = -1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'Accuracy score {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision score {precision_score(y_test, y_pred):.2f}')
print(f'Recall score {recall_score(y_test, y_pred):.2f}')
print(f'Report \n {classification_report(y_test, y_pred)}')
```

```python
param_dist_logreg = {
    "penalty" : ["l1", "l2", "elasticnet"],
    "C" : stats.loguniform(0.0001, 100),
    "l1_ratio" : stats.uniform(0, 1),  # fixed here
    "max_iter" : [300, 500, 700, 1000],
    "tol" : [0.0001, 0.001, 0.01]
}
```

```python
%%time
cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)
random_search = RandomizedSearchCV(estimator = model, param_distributions = param_dist_logreg,
                                  n_iter = 150, scoring = "f1_macro", n_jobs = -1, 
                                  cv = cv, verbose =2, random_state = 42)
random_search.fit(X_train, y_train)
```

```python
best_score = random_search.best_score_
best_params = random_search.best_params_ 
 
```

```python
print(f"Best Score {best_score}")
print(f"Best Parameters {best_params}")
```

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(f'Accuracy score {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision score {precision_score(y_test, y_pred):.2f}')
print(f'Recall score {recall_score(y_test, y_pred):.2f}')
print(f'Report \n {classification_report(y_test, y_pred)}')
```

```python
cm = confusion_matrix(y_test, y_pred)
matrix = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = le.classes_)
fig, ax = plt.subplots()
matrix.plot(ax = ax)
ax.set_title('Confusion Matrix', fontsize = 20, pad = 10)
plt.tight_layout()
plt.show()
```

```python
def test_model(text):
    text = clean_text(text)
    # Vectorize
    X = vectorizer.transform([text])
    x_reduced = svd.transform(X)
    prediction = best_model.predict(x_reduced)
    proba = model.predict_proba(x_reduced)[0]
    confidence = proba.max()  # take highest probability
    # Output result
    if prediction == 1:
        label = "Positive Comment"
    else:
        label = "Negative Comment"
    
    if confidence:
        print(f"{label} | Confidence: {confidence:.2f}")
    else:
        print(label)
```

```python
test_model("The movie was absolutely fantastic, I loved it!")
test_model("It was okay, nothing special.")
test_model("Terrible storyline and poor acting.")
```
